> 5: Il s'agit maintenant de réafficher la matrice toutes les secondes. (Avec la solution)

Le module `datetime` permet de récuperer les heures, les minutes et les secondes facilement en types `int`comme ceci:

```
>>> from datetime import datetime
>>> x = datetime.now()
>>> h, m, s = x.hour, x.minute, x.second
>>> print(h, m, s)
9 51 56
```

Le module `time`possède la méthode `sleep(s)` qui permet de marquer une pause de durée `s` où `s` est en secondes. 

`````python
from datetime import datetime
from time import sleep
def display_mat_by_sec(M, duree):
    sec = datetime.now().second
    while datetime.now().second - sec < duree:
        display_mat(M)
        print()
        sleep(1)
`````
